```python
#importing the dependency
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

```python
# loading the dataset into pandas dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# Head function will give you first five row
# Tail function will give you last five row
credit_card_data.head()
```

```python
# print last five row using tail function
credit_card_data.tail()
```

```python
# To Provide some important information about dataset
credit_card_data.info()
```

```python
# Checking the How many number of missing value in each colom if avl then we have to replace them into meaning full number
credit_card_data.isnull().sum()
```

```python
# Divide the fraud credit card and Normal transaction
# 0 represent Normal or good transaction
# 1 Represent fraud transaction

credit_card_data['Class'].value_counts()
```

The data is highly unbalanced
0-> Normal Transaction
1-> Fraud Transaction

```python
# Seperating the data between fraud and Normal
normal = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
# Printing the normal data and fraud data
print(normal.shape)
print(fraud.shape)
```

```python
#statistical measure of the data
normal.Amount.describe()
```

```python
#statistical measure of the data
# It is not percent it is percentile
# It is mean or average value
fraud.Amount.describe()
```

```python
# Compare the Value for both transaction
# By using this difference our Machine Learning Model will predict whether card is fraud or not
# There is big difference betweem 0 and 1
credit_card_data.groupby('Class').mean()
```

:Under sampling
built a sample data set thet contain similar disrtibution and normal transaction

Number of Fradu Transaction is 492

```python
normal_sample = normal.sample(n=492)
```

```python
# Now Concatinante two Data frame
new_dataset = pd.concat([normal_sample, fraud], axis=0)
```

```python
# Print first five row
# This is the normal sample
new_dataset.head()
```

```python
# Print last five row
# This is the fraud sample
new_dataset.tail()
```

```python
# 0 Represent Normal Transation
# 1 Represent Fraud Transation
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# Splitting the data into feature and target
# means we deleted class cloumn ans remaining thing store into X variable
# and we store class coloun into Y variable, X has no class column
# and Y has only class coloum
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
# X has a feature except class
print(X)
```

```python
# it contain the labels , Y has a class
print(Y)
```

```python
# Split the data into training data set adn testing data set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
# 984 value is the Original data set
#787 is the 80% of data for traning
#197 is the 20% of data for testing
print(X.shape, X_train.shape, X_test.shape)
```

Model Traning Logistic Regression

```python
# we will store Logistic Regression model into model variable
model = LogisticRegression()
```

```python
# Traning the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

```python
# Model Evaluation
# Accuracy score
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score on traning data : ', training_data_accuracy)
# we got 94.1 % of accuracy score which is very good
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on traning data : ', test_data_accuracy)
```
